```python
import numpy as np
import matplotlib.pyplot as plt

dates = np.array([1, 2, 3, 4, 5, 6, 7])
angular_positions = np.array([10, 25, 40, 55, 70, 85, 100])

angular_displacements = np.diff(angular_positions)
average_displacement_per_day = np.mean(angular_displacements)

rotation_period = 360 / average_displacement_per_day

print("Average Angular Displacement per Day: {:.2f} degrees".format(average_displacement_per_day))
print("Esitmated Solar Rotation Period: {:.2f} days".format(rotation_period))

plt.plot(dates, angular_positions, marker='o', linestyle= '-', color= 'purple')
plt.title("Sunspot Angular Positions Over time")
plt.xlabel("Days")
plt.ylabel("Angular Position (degrees)")
plt.grid(True)
plt.show()
```
