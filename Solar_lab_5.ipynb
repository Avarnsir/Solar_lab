```python
import numpy as np
import matplotlib.pyplot as plt
# Constants
sigma = 5.670e-8  # Stefan-Boltzmann constant (W/m^2K^4)
R_earth = 6.371e6  # Earth's radius in meters
A = np.pi * R_earth**2  # Cross-sectional area of Earth
T_sun = 5778  # Effective temperature of Sun in Kelvin
t = 1  # Time of exposure in seconds

# Compute Q using Stefan-Boltzmann Law
Q = sigma * A * T_sun**4 * t

# Average Earth-Sun distance
d = 1.496e11  # meters

# Compute Solar Constant (S)
S = (Q * d**2) / (A * t)

# Print results
print(f"Heat Energy Absorbed by Earth (Q): {Q:.2e} J")
print(f"Solar Constant (S): {S:.2f} W/m²")
```
